# Preparation

```python
import pandas as pd
```

```python
df_mortality = pd.read_excel(io='MortalityDataWHR2021C2.xlsx')
```

```python
df_happiness = pd.read_excel(io='DataForFigure2.1WHR2021C2.xls')
```

```python
df_regions = df_happiness[['Country name', 'Regional indicator']]
```

```python
df = df_regions.merge(df_mortality)
```

```python
df.head()
```

# Islands

## Number of Islands 

```python
df.Island.sum()
```

## Which region had more Islands?

```python
df.groupby('Regional indicator').Island.sum()
```

## Show all Columns for these Islands

```python
mask_region = df['Regional indicator'] == 'Western Europe'
```

```python
mask_island = df['Island'] == 1
```

```python
df_europe_islands = df[mask_region & mask_island]
```

```python
df_europe_islands
```

## Mean Age of across All Islands?

```python
df_europe_islands['Median age'].mean()
```

# Female Heads of State

## Number of Countries with Female Heads of State 

```python
df['Female head of government'].sum()
```

## Which region had more Female Heads of State?

```python
df.groupby('Regional indicator')['Female head of government'].sum().sort_values(ascending=False)
```

## Show all Columns for these Countries

```python
mask_region = df['Regional indicator'] == 'Western Europe'
```

```python
mask_female = df['Female head of government'] == 1
```

```python
df_europe_femaleheads = df[mask_region & mask_female]
```

```python
df_europe_femaleheads
```

## Mean Age of across All Countries?

```python
df_europe_femaleheads['Median age'].mean()
```
